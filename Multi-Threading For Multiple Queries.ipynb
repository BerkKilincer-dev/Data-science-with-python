```python
import pandas as pd
import threading
import time

# İşlenecek dosya sayısını tanımlar.
num_files=12

# Filtrelenmiş kişileri depolamak için boş bir liste oluşturur (çoklu iş parçacığı için).
result2=[]

# Her bir dosya için veri işleme fonksiyonunu tanımlar.
def data_parser(filename):
    try:
        # JSON dosyasını bir DataFrame olarak okur.
        df=pd.read_json(filename)
        # 'Dogum Tarihi' sütununu datetime formatına dönüştürür.
        df['Dogum Tarihi'] = pd.to_datetime(df['Dogum Tarihi'])
        # Kadıköy ilçesinden ve 1990-01-01 sonrası doğan kişileri filtreler.
        kadikoy_2000=df[(df['Ilce']=='KADIKÖY') & (df['Dogum Tarihi']>'1990-01-01')]
        # Soyadında 'ÖZ' geçen kişileri filtreler.
        oz_kadikoy_2000=kadikoy_2000.loc[kadikoy_2000['soyisim'].str.contains('ÖZ')]
        # Filtrelenmiş veriyi bir sözlük listesine dönüştürür.
        data=oz_kadikoy_2000.to_dict(orient='records')
        # Her kişiyi result2 listesine ekler.
        for kisi in data:
            result2.append(kisi)
    except Exception as e:
        # Hata durumunda istisna mesajını ve dosya adını yazdırır.
        print(f'istisna var: {filename}')


# Zaman ölçümü başlangıcı.
start = time.time()

# Dosya adlarını depolamak için boş bir liste oluşturur.
databases=[]

# 1'den num_files'e kadar dosya adlarını oluşturur ve listeye ekler.
for i in range(1,(num_files+1)):
    filename=f'contactlist2_{i}.json'
    databases.append(filename)

# İş parçacıklarını depolamak için boş bir liste oluşturur.
threads=[]

# Her bir veritabanı için bir iş parçacığı oluşturur ve listeye ekler.
for db in databases:
    th = threading.Thread(target=data_parser, args=(db,))
    threads.append(th)

# Tüm iş parçacıklarını başlatır.
for th in threads:
    th.start()

# Tüm iş parçacıklarının bitmesini bekler.
for th in threads:
    th.join()

# Zaman ölçümü bitişi.
end = time.time()
# Çalışma süresini yazdırır.
print(f'Current Execution Time:{end-start}')
# result2 listesinin toplam uzunluğunu yazdırır.
print(len(result2))
```

```python
import pandas as pd
import time
import concurrent.futures


# İşlenecek dosya sayısını tanımlar.
num_files=12

# Filtrelenmiş kişileri depolamak için boş bir liste oluşturur.
result3=[]

# Her bir dosya için veri işleme fonksiyonunu tanımlar.
def data_parser(filename):
    try:
        # JSON dosyasını bir DataFrame olarak okur.
        df=pd.read_json(filename)
        # 'Dogum Tarihi' sütununu datetime formatına dönüştürür.
        df['Dogum Tarihi'] = pd.to_datetime(df['Dogum Tarihi'])
        # Kadıköy ilçesinden ve 1990-01-01 sonrası doğan kişileri filtreler.
        kadikoy_2000=df[(df['Ilce']=='KADIKÖY') & (df['Dogum Tarihi']>'1990-01-01')]
        # Soyadında 'ÖZ' geçen kişileri filtreler.
        oz_kadikoy_2000=kadikoy_2000.loc[kadikoy_2000['soyisim'].str.contains('ÖZ')]
        # Filtrelenmiş veriyi bir sözlük listesine dönüştürür.
        data=oz_kadikoy_2000.to_dict(orient='records')
        # Her kişiyi result3 listesine ekler.
        for kisi in data:
            result3.append(kisi)
    except Exception as e:
        # Hata durumunda istisna mesajını ve dosya adını yazdırır.
        print(f'istisna var: {filename}')


# Zaman ölçümü başlangıcı.
start = time.time()

# Dosya adlarını depolamak için boş bir liste oluşturur.
databases=[]

# 1'den num_files'e kadar dosya adlarını oluşturur ve listeye ekler.
for i in range(1,(num_files+1)):
    filename=f'contactlist2_{i}.json'
    databases.append(filename)

# concurrent.futures.ThreadPoolExecutor kullanarak maksimum 4 iş parçacığı ile paralel işlem yapar.
with concurrent.futures.ThreadPoolExecutor(max_workers=4) as executor:
    # Her bir veritabanı için data_parser fonksiyonunu bir iş parçacığına gönderir.
    futures = [executor.submit(data_parser, db) for db in databases]
    # Tüm iş parçacıklarının tamamlanmasını bekler ve oluşabilecek istisnaları yakalar.
    for future in concurrent.futures.as_completed(futures):
        future.result()  # İş parçacığı yürütülmesi sırasında oluşan herhangi bir istisnayı yükseltir.

# Zaman ölçümü bitişi.
end = time.time()
# Çalışma süresini yazdırır.
print(f'Current Execution Time: {end - start}')
```
