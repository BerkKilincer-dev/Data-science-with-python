```python
from bs4 import BeautifulSoup
```

```python
#imdb.html parser dosyasini BeautifulSoup fonksiyonu ile ac
#soup referansini olustur

with open('imdb.html') as f:
    soup=BeautifulSoup(f,'html.parser')
soup
```

```python
# td tag'li olan verilerden class titleColumn olan verileri cek
#filmler variable'ina ata
filmler=soup.find_all('td',{'class':'titleColumn'})
filmler
```

```python
#filmler listesinin 0.index'i
filmler[0]
```

```python
#filmler listesinin 0.index'inin text'i
filmler[0].text
```

```python
#filmler listesinin 0.index'inin text'ini split et: \n
filmler[0].text.split('\n')
```

```python
#bos imdb dict variable olustur
#bos isim list variable olustur
#bos yil list variable olustur
#filmler icin for loop'u icinde
#her bir array'in text'inin \n split'i icinde index 2'leri isim listesine gonder
#her bir array'in text'inin \n split'i icinde index 3'leri yil listesine gonder
#yil daki ( ve ) stringlerini sil ve int'e cevir
#imdb dictionary'sinin 'isim' key'inin value degeri: isim listesi
#imdb dictionary'sinin 'yil' key'inin value degeri: yil listesi
#imdb dictionary variable print et

imdb={}
isim=[]
yil=[]
for film in filmler:
    isim.append(film.text.split('\n')[2].strip())
    year=film.text.split('\n')[3].replace('(','')
    year=year.replace(')','')
    yil.append(int(year))
imdb['isim']=isim
imdb['yil']=yil
imdb
```

```python
# td tag'li olan verilerden class ratingColumn imdbRating olan verileri cek
#puanlar variable'ina ata
puanlar = soup.find_all('td', {'class' : 'ratingColumn imdbRating'})
puanlar
```

```python
#puanlar listesindeki her bir array'in text'ini for loopunda print et
for puan in puanlar:
    print(puan.text)
```

```python
#puanlar listesindeki her bir array'in text'ini strip edip, for loopunda print et
for puan in puanlar:
    print(puan.text.strip())
```

```python
#bos reyting_puan listesi olustur
#puanlar listesindeki her bir array'in text'ini for loopunda float'a cevirip
#reyting_puan listesine ekle
reyting_puan=[]
for puan in puanlar:
    reyting_puan.append(float(puan.text.strip()))
reyting_puan
```

```python
#imdb dictionary'sinin 'puan' key'inin value degeri: reyting_puan listesi
#imdb dictionary variable print et
imdb['puan']=reyting_puan
imdb
```

```python
#imdb dictionary'isinden dataframe olustur
#df variable
import pandas as pd 
df =pd.DataFrame.from_dict(imdb)
df
```

```python
#dataframe info al
df.info()
```

```python
#en eski yildan bugune filmleri sort et
df_yil_gecmisten_bugune=df.sort_values('yil',ascending=True)
df_yil_gecmisten_bugune
```

```python
#en guncel yildan en eskiye filmleri sort et
df_yil_2022_gecmise=df.sort_values('yil',ascending=False)
df_yil_2022_gecmise
```

```python
# bos imdb_liste listesi olustur
#filmler listesinin uzunlugunun range'inde
#imdb_liste'ye dictionary ekle
#eklenen dictionary de: 'film adi' key'inin value degeri isim listesinin iteration'i
#eklenen dictionary de: 'film yili' key'inin value degeri yil listesinin iteration'i
#eklenen dictionary de: 'film puani' key'inin value degeri reyting_puan listesinin iteration'i
#imdb_liste print et

imdb_liste=[]
for i in range(len(filmler)):
    imdb_liste.append({'film adi':isim[i],
                      'film yili':yil[i],
                       'film puani':reyting_puan[i]})
imdb_liste
```

```python
#imdb_liste den dataframe olustur
df2=pd.DataFrame(imdb_liste)
df2
```

```python
#imdb_sonuc.txt olustur
#imdb_liste listesindeki her dictionary icin:
#'isim: '+filmin adi+' yil: '+filmin yili + ' Puan:' + filmin puani
#yazdir
with open('imdb_sonuc.txt','w') as f:
    for film in imdb_liste:
        f.write('isim:'+film['film adi']+'yil: '+str(film['film yili'])+'Puan:'+str(film['film puani'])+'\n')
```
