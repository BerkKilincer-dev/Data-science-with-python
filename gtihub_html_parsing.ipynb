```python
#url sayfasinin html verisini print et 
import requests
url='https://github.com/ErdenizUnvan?tab=repositories'
response=requests.get(url)
print(response.text)
```

```python
#beautifulsoup kutuphanesinden BeautifulSoup fonksiyonunu import et
#url sayfasinin html verisini BeautifulSoup fonksiyonu ile parselle

from bs4 import BeautifulSoup
soup=BeautifulSoup(response.text,'html.parser')
soup
```

```python
html datasinda github repository isimleri ve linkleri nerede? 
<li class="col-12 d-flex flex-justify-between width-full py-4 border-bottom color-border-muted public source"
```

```python
#soup variable'inda yukardaki verinin oldugu butun datalari bul
#veriler variable
veriler=soup.find_all('li',{'class':'col-12 d-flex flex-justify-between width-full py-4 border-bottom color-border-muted public source'})
veriler
```

```python
veriler variable'i icindeki her bir data icinde url link ve metin nerede

<h3 class="wb-break-all">
```

```python
#veriler variable'i icindeki her bir data icinde url link ve metin nerede
for veri in veriler:
    print(veri.find('h3',{'class':'wb-break-all'}))
```

```python
#repository isimleri=> metinler
for veri in veriler:
    print(veri.find('h3',{'class':'wb-break-all'}).text.strip().split('Public')[0].strip())
```

```python
#url link iÃ§in 
for veri in veriler:
    link=veri.find('h3',{'class':'wb-break-all'}).a['href']
    print(link)
```

```python
#liste icinde dictionary olustur
#dictionary'de metin key'inin value degeri icin repository baslik metni
#dictionary'de url key'inin value degeri icin repository url'i

git_hub=[]
content=[]
for veri in veriler:
    metin=veri.find('h3',{'class':'wb-break-all'}).text.strip().split('Public')[0].strip()
    link=veri.find('h3',{'class':'wb-break-all'}).a['href']
    url='https://github.com'+link
    content={'metin':metin,'url':url}
    git_hub.append(content)
git_hub
```
