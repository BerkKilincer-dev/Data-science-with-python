```python
import requests
from bs4 import BeautifulSoup
result=[]
for i in range(1,11):
    url=f"https://stackoverflow.com/questions/tagged/python?tab=newest&page={i}&pagesize=50"
    response=requests.get(url)
    soup=BeautifulSoup(response.text,'html.parser')
    data=soup.find_all('div',{'class':'s-post-summary--content'})
    for veri in data:
        link='https://stackoverflow.com'+veri.find('a',{'class':'s-link'})['href']
        metin=veri.find('a',{'class':'s-link'}).text
        content={'soru':metin,
            'link':link}
        result.append(content)
    print(len(result))
#result
```
